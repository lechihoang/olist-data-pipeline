```python
# Import PySpark libraries
from pyspark.sql import SparkSession

# Create or get Spark session
spark = SparkSession.builder \
    .appName("Olist Data Analysis") \
    .config("spark.driver.memory", "4g") \
    .getOrCreate()

print("Spark session created successfully!")
print(f"Spark version: {spark.version}")
```

```python
# Cell 1: Cấu hình
# ==================================

# 1. Đường dẫn file CSV 
# Update path to match your workspace structure
raw_csv_path = "/Users/tranthihieu/Documents/Brazil Ecommere/data/olist_orders_dataset.csv"

# In ra để kiểm tra
print(f"File CSV nguồn: {raw_csv_path}")
```

```python
# Cell 2: Đọc và Hiển thị (Thử nghiệm)
# ==================================

# Dùng Spark (biến 'spark' đã có sẵn) để đọc file CSV
# - "header=True": Báo cho Spark biết dòng đầu tiên là tên cột
# - "inferSchema=True": Bảo Spark tự động đoán kiểu dữ liệu
df = (spark.read
    .format("csv")
    .option("header", "true")
    .option("inferSchema", "true") 
    .load(raw_csv_path)
)

print("Đọc file thành công! Đang hiển thị dữ liệu...")

# Dùng lệnh 'display' (đã có sẵn) để hiển thị bảng
display(df)
```
